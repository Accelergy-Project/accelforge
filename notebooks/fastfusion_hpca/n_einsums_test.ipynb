```python
# Imports
import fastfusion as ff
from IPython.display import SVG, display
import os
ff.set_n_parallel_jobs(32)


# Initialize the specification and show the workload.
BATCH_SIZE = 4
N_TOKENS = 4096
FUSE = True

spec = ff.Spec.from_yaml(
    # "arches/tpu_v4i_like.arch.yaml",
    "../../examples/arches/tpu_v4i_like.arch.yaml",
    # "arches/simple.arch.yaml",
    "../../examples/workloads/matmuls.workload.yaml",
    # "workloads/matmuls8_mixed.workload.yaml",
    jinja_parse_data=dict(
        M=8192,
        KN=8192,
        N_EINSUMS=256,
    )
)
spec.mapper.ffm.metrics |= ff.mapper.Metrics.LATENCY | ff.mapper.Metrics.ENERGY

# If fusion is disabled, keep all tensors in main memory.
if not FUSE:
    spec.arch.nodes["MainMemory"].constraints.tensors.keep = "All()"

# display(SVG(spec.workload.render()))
```

```python
pmappings = ff.mapper.FFM.make_pmappings(spec)
joined = ff.mapper.FFM.join_pmappings(spec, pmappings)
```
